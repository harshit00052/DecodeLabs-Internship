```python
# Run this in a Colab cell to set up SQLite from your Excel file
import pandas as pd
import sqlite3

df = pd.read_csv('/content/cleaned_dataset.csv')
conn = sqlite3.connect('orders.db')
df.to_sql('orders', conn, if_exists='replace', index=False)
```

```python
def run_query(sql):
    result = pd.read_sql_query(sql, conn)
    print(result.to_string(index=False))
    return result
```

***SELECT Queries***

```python
run_query("SELECT * FROM orders LIMIT 5")
```

The orders table has 14 columns including OrderID, Date, CustomerID,
Product, Quantity, UnitPrice, PaymentMethod, OrderStatus, and TotalPrice.
The dataset contains 1,200 orders spanning from 2023 to 2025.

```python
run_query("""SELECT OrderID, Product, TotalPrice, OrderStatus
FROM orders
LIMIT 5""")
```

Selecting key columns shows a mix of products and statuses across orders.
Not all orders are successful — Cancelled and Returned statuses appear
alongside Shipped and Delivered ones.

 ***Filter with WHERE***

```python
run_query("""SELECT * FROM orders
WHERE OrderStatus = 'Delivered'""")
```

Only 231 out of 1,200 orders are Delivered — just 19.25% of all orders.
The majority of orders are still Pending, Shipped, Cancelled, or Returned,
which means the business is successfully completing less than 1 in 5 orders.

```python
run_query("""SELECT OrderID, CustomerID, Product, TotalPrice
FROM orders
WHERE PaymentMethod = 'Credit Card' AND OrderStatus = 'Cancelled'""")
```

Several Credit Card orders were cancelled. This is financially damaging
because cancelled Credit Card transactions may involve refund processing
costs. The business should investigate the reason behind these cancellations
to reduce revenue loss.

```python
run_query("""SELECT OrderID, Date, Quantity, UnitPrice, TotalPrice
FROM orders
WHERE Product = 'Laptop'""")
```

Laptop unit prices range widely — from as low as $59.55 to $664.42 —
suggesting multiple laptop models at different price points. Some customers
purchase up to 5 units in a single order, indicating possible bulk or
business purchases.

***Sort with ORDER BY***

```python
run_query("""SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10""")
```

The highest value order is $3,456.40 for a Tablet. All top 10 orders
exceed $3,300 and include Tablets, Monitors, Laptops, Chairs, and Printers.
These high-value customers should be targeted for loyalty and retention
programs.

```python
run_query("""SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice ASC
LIMIT 10""")
```

The lowest value order is just $11.39 for a Phone. The bottom 10 orders
are all under $30, likely representing single-unit purchases of budget
items or heavily discounted products using coupon codes.

```python
run_query("""SELECT OrderID, Date, Product, TotalPrice
FROM orders
WHERE OrderStatus = 'Delivered'
ORDER BY Date DESC""")
```

The most recent delivered order was on June 28, 2025 — a Phone worth
$548.24. Delivered orders include all product types with no single product
dominating, showing balanced delivery performance across categories.

***Aggregate with COUNT, SUM, AVG***

```python
run_query("SELECT COUNT(*) AS TotalOrders FROM orders")
```

The dataset contains exactly 1,200 orders. This is a solid sample size
for drawing reliable business insights across all product categories,
payment methods, and order statuses.

```python
run_query("SELECT SUM(TotalPrice) AS TotalRevenue FROM orders")
```

Total gross revenue across all 1,200 orders is $1,264,761.96. This
includes cancelled and returned orders, so actual collected revenue
is significantly lower than this figure.

```python
run_query("SELECT AVG(TotalPrice) AS AverageRevenue FROM orders")
```

The average order value is $1,053.97. This relatively high average is
likely driven by customers purchasing multiple units per order of
electronics and furniture items.

```python
run_query("""SELECT SUM(TotalPrice) AS DeliveredRevenue
FROM orders
WHERE OrderStatus = 'Delivered'""")
```

Only $242,600.32 out of $1,264,761.96 total revenue comes from Delivered
orders — just 19.2%. Over 80% of gross revenue is tied up in Pending,
Shipped, Cancelled, or Returned orders, which represents a significant
business risk.

***GROUP BY***

```python
run_query("""SELECT OrderStatus, COUNT(*) AS OrderCount
FROM orders
GROUP BY OrderStatus
ORDER BY OrderCount DESC""")
```

Result:

Cancelled orders are the highest at 250, and Delivered orders are the
lowest at 231. The fact that more orders are cancelled than delivered is

***HAVING***

```python
run_query("""SELECT Product, SUM(TotalPrice) AS TotalRevenue
FROM orders
GROUP BY Product
HAVING SUM(TotalPrice) > 50000
ORDER BY TotalRevenue DESC""")
```

All 7 products exceed $50,000 in total revenue, confirming every product
category is commercially strong. Chair leads at $195,620 and Phone is
lowest at $151,722 — but even the lowest product generates over 3x the
threshold, showing a healthy product lineup.

```python
run_query("""SELECT PaymentMethod, COUNT(*) AS UsageCount
FROM orders
GROUP BY PaymentMethod
HAVING COUNT(*) > 200""")
```

Result:
       Cash  246
Credit Card  234
 Debit Card  232
  Gift Card  230
     Online  258

All 5 payment methods are used more than 200 times each. Online is the
most frequently used at 258 transactions and Gift Card the least at 230.
The near-equal usage across all methods shows customers have no strong
preference — the business should continue supporting all payment options.
